# Task and procedure

The task consists in replicating the variable price using other features included in the dataset. The only constraint is to use at least 5 of them and being able to explain which ones are the most relevant to replicate the variable price.

The procedure I will use for this task is the following:

1) I will try different machine learning models and I will pick the one that is ensuring the best fit in terms of loss functions; I will use Rsquared, mean squared error and mean absolute error as criteria to select the model.

2) After having selected the best performing model in terms of loss functions I will use the function feature importance of sklearn to have a ranking of the most important features for the fit. I will then delete the non-useful variable and re-train the model with only the selected features that are minimum five. 

In this way I will let the best model to do vriable selection.

# Considerations

This task is not intended to create a model able to predict the variable price (as I also asked by email), so we do not care if the model is overfitting; on the contrary, if the model overfits it is even better, because we have a lower loss fuction and the charts of the real time series and generated one will be almost coincident.
Having made these considerations is not relevant if our model is interpretable or not, what it matters is that we can fit the variable price using the other variables and that we can point out which features are the most important for the fit and perform a variable selection without losing precision.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize= (20,20))
sns.heatmap(df[df.columns[1:]].astype('float64').corr(), cmap="YlGnBu", annot=True)
```

I tried diffferent machine learning models as random forest, decision tree regressor, multilayer perceptron and gradient boosting that is the best performer. Considering the criteria, the model I picked is gradient bossting and in the following piece of code you can find all the info requested in the task.

First I delete the column IK1 Comdty that is almost always NaN and so is not relevant to fit the variable price.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
import numpy as np
import matplotlib.pyplot as plt

df = df.drop(columns='IK1 Comdty')

x = df[df.columns[2:]]
y = df['price']
```

Now I fit the best model I was able to find: Gradint Boosting Regressor and considering I don't care about overfitting I train it with a number of estimators = 15000 that is 15 times the default value to avoid overfitting. This will imply a longer training time but in the task no penalty is specified in case of longer training time.

```python
gb = GradientBoostingRegressor(random_state=0, n_estimators = 15000)

gb.fit(x, y)


score2 = gb.score(x, y)
ypred2 = gb.predict(x)


mse2 = mean_squared_error(y, ypred2)
mae2 = mean_absolute_error(y,ypred2)

print("R2:{1:.4f}, MSE:{2:.4f}, MAE:{2:.4f}, RMSE:{3:.4f}".format(score2, mse2, mae2, np.sqrt(mse2)))
```

Below I show the chart with the true time series and the one generated to fit the original one:
(are quite difficult to distinguish below so I will do other 2 plots where they are distinct)

```python
plt.figure(figsize=(20,7))
plt.plot(df['Dates'], gb.predict(x))
plt.legend('Generated')
plt.plot(df['Dates'],y)
plt.legend('True')
plt.title('True vs Generated')
```

Below I show the generated one first and then the true one:

```python
plt.figure(figsize=(20,7))

plt.subplot(1,2,1)
plt.plot(df['Dates'], gb.predict(x))
plt.legend('Generated')
plt.title('Generated')

plt.subplot(1,2,2)
plt.plot(df['Dates'],y, color = 'g')
plt.legend('True')
plt.title('True')
```

In the last chart below I show the final answer to the task: feature importance; as you can observe from the chart my model has ranked all the features based on how useful and used they are during the train and fit. In this way it is very simple to select the ones that are useful and discard the ones that are not.
The last 5 variables in the chart have an importance close to 0 and if discarded the model would perform almost exactly the same. On the contrary the first three variables are the ones that are fundamental to explain the performance of the model and to replicate the variable price.

```python
importances = gb.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(20,15))
ax.barh(range(len(importances)), importances[indices])
ax.set_yticks(range(len(importances)))
_ = ax.set_yticklabels(np.array(x.columns)[indices])
plt.title('Feature Importance')
```

The task was to replicate the variable price only from the selected features so below I will re-perform the train deleting the variables with importance null or close to null.

```python
df_filtered = df.drop(columns = ['UKX Index', 'F90503M Index', 'EURUSD Curncy', 'TY1 Comdty',
       'HKD Curncy', 'DAX Index', 'EURZAR Curncy', 'DXY Curncy',
       'IBOV Index', 'KOSDAQ Index', 'ASX Index'])

x_new = df_filtered[df_filtered.columns[2:]]
y = df_filtered['price']
```

```python
gb = GradientBoostingRegressor(random_state=0, n_estimators = 15000)

gb.fit(x_new, y)


score2 = gb.score(x_new, y)
ypred2 = gb.predict(x_new)


mse2 = mean_squared_error(y, ypred2)
mae2 = mean_absolute_error(y,ypred2)

print("R2:{1:.4f}, MSE:{2:.4f}, MAE:{2:.4f}, RMSE:{3:.4f}".format(score2, mse2, mae2, np.sqrt(mse2)))
```

As you can see the model is still able to perform incredibly good but the loss is slightly higher than before, when we were using all variables.
You can choose if to consider the model with all the variables or this one, their performance is very close; the reason why now the model is performing not as good as before is that the model itself is able to perform a variable selection.

```python
plt.figure(figsize=(20,7))
plt.plot(df['Dates'], gb.predict(x_new))
plt.legend('Generated')
plt.plot(df['Dates'],y)
plt.legend('True')
plt.title('True vs Generated')
```
